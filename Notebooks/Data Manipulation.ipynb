## Chapter 1: What is Pandas?

Pandas is a Python library originally developed as a high performance, flexible framework to perform quantitative analysis on financial data

Data can be loaded into Python using pandas from following sources:
* Flat files
* Parquet files
* SQL databases (including TeraData)
* JSON
etc

```python
## Loading Data in Pandas

data_frame = pd.read_csv("Assets/dataset_diabetes/dataset_diabetes/diabetic_data.csv", na_values='?')
```

```python
data_frame.head()
```

## Chapter 2: Working with Data-Frames

```python
##List of columns in a data_frame
data_frame.columns
```

```python
#Types of columns in the data_frame
data_frame.info()
```

```python
#Data Description for numerical data types
data_frame.describe()
```

```python
#Selecting one column
admission_type_id = data_frame["admission_type_id"]
```

```python
# admission_type_id.values
# admission_type_id.value_counts
```

```python
#Selecting multiple columns
sub_frame = data_frame[["patient_nbr","admission_type_id", "time_in_hospital"]]
sub_frame
```

```python
#Conditional subset

data_frame.loc[data_frame["admission_type_id"] == 3]
```

```python
#Multi Conditional subset

data_frame.loc[(data_frame["admission_type_id"] == 3)&(data_frame["race"] == "Caucasian")]
```

```python
#Iterating over dataset
for row_number, row_data in data_frame[:10].iterrows():
    print(row_data["patient_nbr"], row_data["insulin"])
```

## Chapter 3 : SQL Like Operations

```python
#Group By
data_group = data_frame.groupby("gender")
```

```python
data_group.first()
```

```python
data_frame.loc[data_group.groups['Male'].values]
# Read More: https://realpython.com/pandas-groupby/
```

```python
#Join

admission_types = pd.read_excel("Assets/dataset_diabetes/dataset_diabetes/admission_types.xlsx")
```

```python
pd.merge(data_frame, admission_types, on="admission_type_id")
```
